```python
from llama_cpp import Llama

my_model_path = "/home/anush/Desktop/LLMs/llama3/Llama-3-Instruct-8B-SPPO-Iter3-Q6_K.gguf"
CONTEXT_SIZE = 2048

model = Llama(model_path=my_model_path, n_ctx=CONTEXT_SIZE)

def generate_text_from_prompt(user_prompt, max_tokens = 1024, temperature = 0.7, top_p = 0.1, echo = True, stop = ["Q", "\n"]):
    model_output = model(
       user_prompt,
       max_tokens=max_tokens,
       temperature=temperature,
       top_p=top_p,
       echo=echo,
       stop=stop,)
    return model_output


if __name__ == "__main__":
    system_prompt="you are a helpful chatbot"
    pmpt=
""" 
    
AutoGen is an open-source framework from Microsoft Research that allows developers to build AI agents and 
facilitate cooperation between multiple agents to solve tasks. It provides a multi-agent conversation framework 
as a high-level abstraction, and features include: Agents that can converse with other agents, LLM and tool use support, 
Autonomous and human-in-the-loop workflows, and Multi-agent conversation patterns.

tell me what are the features of autogen?

"""
    prompt = f"""
    
<|begin_of_text|><|start_header_id|>system<|end_header_id|>

{system_prompt}<|eot_id|><|start_header_id|>user<|end_header_id|>

{pmpt}<|eot_id|><|start_header_id|>assistant<|end_header_id|>

"""
    model_response = generate_text_from_prompt(prompt)
    final_result = model_response["choices"][0]["text"].strip()
    print(model_response)
    print(f'\n\n {final_result}')
```
