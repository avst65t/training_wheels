### pdf generation

```python
from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, PageBreak
import re

# Constants for margins
PAGE_WIDTH, PAGE_HEIGHT = A4
MARGIN_LEFT = 60
MARGIN_RIGHT = 60

# Read content from file
with open("final_content.txt", "r", encoding="utf-8") as file:
    txt = file.read()


def format_bold(text):
    """ Convert **bold** to ReportLab's <b>bold</b> format """
    return re.sub(r"\*\*(.*?)\*\*", r"<b>\1</b>", text)

def create_pdf(filename):
    doc = SimpleDocTemplate(filename, pagesize=A4, leftMargin=MARGIN_LEFT, rightMargin=MARGIN_RIGHT)

    styles = getSampleStyleSheet()
    normal_style = styles["Normal"]
    normal_style.leading = 14  # Adjust line spacing
    bold_style = ParagraphStyle("Heading", parent=styles["Heading2"], fontSize=14, spaceAfter=10)

    elements = []

    # Page 1: Disclaimer
    elements.append(Paragraph("Disclaimer", bold_style))
    elements.append(Paragraph(format_bold(discl).replace("\n", "<br/>"), normal_style))
    elements.append(PageBreak())  # Force new page

    # Page 2: Introduction
    elements.append(Paragraph("Introduction", bold_style))
    elements.append(Paragraph(format_bold(intro).replace("\n", "<br/>"), normal_style))
    elements.append(PageBreak())  # Force new page

    # Page 3: Table of Contents
    elements.append(Paragraph("Table of Contents", bold_style))
    elements.append(Paragraph(format_bold(report_structure).replace("\n", "<br/>"), normal_style))
    elements.append(PageBreak())  # Force new page

    # Page 4+: Main Report Content
    elements.append(Paragraph("COMPREHENSIVE HEALTH ANALYSIS PREPARED FOR JEAN YVES CHRISTIAN SIREAU", bold_style))
    elements.append(Paragraph(format_bold(txt).replace("\n", "<br/>"), normal_style))

    # Build the PDF
    doc.build(elements)

# Generate the PDF
create_pdf("ai_report_stage_2.pdf")
```
